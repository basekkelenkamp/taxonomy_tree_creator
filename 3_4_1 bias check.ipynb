```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Import data
df = pd.read_csv("input-data/adult-dataset.csv")

# Replace numerical values with labels
df = df.query("income == '>50K'")[['sex']]
df['sex'] = df['sex'].replace({1: 'a', 2: 'b'})

# Plot the distribution of the labeled column
sns.countplot(x='sex', data=df)
plt.title("BIAS check male vs female with income >50K")
plt.xlabel("column_name labels")
plt.show()
```

```python
scenario = "scenario-b-coarse"
scenarioFile = scenario.replace("-", "_")
kArray = [2, 5, 10, 15, 20, 25, 30, 40, 50, 75]

for i, x in enumerate(kArray, start=1):
  print(x)
  df = pd.read_csv(f"output-data/transformed-dataset-3.4/{scenario}/{scenarioFile}_k{x}.csv")

  # Replace numerical values with labels
  df = df.query("income == '>50K'")[['sex']]

  # Plot the distribution of the labeled column
  sns.countplot(x='sex', data=df)
  plt.title(f"Transformed dataset {scenario} BIAS check K{x}")
  plt.xlabel("Male vs Female (with income higher than $50K)")
  
      # Add the value of each bar inside the bar
  for i in range(len(df['sex'].value_counts())):
    count = df['sex'].value_counts()[i]
    plt.annotate(str(count), (i, count), ha = 'center', xytext = (0, 0), textcoords = 'offset points', va = 'top')
  
  plt.savefig(f"output-data/bias-check-male-female-3.4/{scenario}/{x}.png", dpi = 150, bbox_inches='tight')
  plt.show()
```
