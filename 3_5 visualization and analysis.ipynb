```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a dictionary with the values as the keys and the accuracy as the values
accuracy = {2: 50,
            5: 20,
            10: 10,
            15: 6.66667,
            20: 5,
            25: 4,
            30: 3.22581,
            40: 2.38095,
            50: 2,
            75: 1.28205}



# Plot the line chart
sns.lineplot(x=list(accuracy.keys()), y=list(accuracy.values()))

# Add title and labels
title = 'scenario_a_coarse_highest_risk'
plt.title(title)
plt.xlabel("K-anonymity")
plt.ylabel("Highest risk (%)")

# Show and save the plot
plt.savefig(f"output-data/visualization-and-analysis-3.5/{title}.png", dpi = 150, bbox_inches='tight')
plt.show()
```
